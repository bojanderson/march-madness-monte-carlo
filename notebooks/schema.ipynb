```python
import pandas as pd
```

```python
import os

# get current working directory
current_dir = os.getcwd()

# move up one level to parent directory
os.chdir("..")

# move to "data" directory
os.chdir("data")

# confirm new working directory
data_dir = os.getcwd()

files = pd.DataFrame.from_dict({"file": []})


for file in os.listdir(data_dir):
    df = pd.read_csv(file)
    columns = df.columns

    for column in columns:
        if column not in files.columns:
            files[column] = 0

    # Add a new row to the dataframe with the name of the file and 1 for each column
    # that exists in the file and a 0 for each column that does not exist in the file
    files = pd.concat(
        [
            files,
            pd.DataFrame.from_dict(
                {"file": [file], **{column: 1 for column in columns}}
            ),
        ],
        ignore_index=True,
    )

files
```
